```r
x_sample <- rexp(n, rate = 1)

generate_scatter_norm <- function(n=40, beta_0 = 5, beta_1 = -2) {
  #x_sample <- rexp(n, rate = 1)
  e_sample <- rnorm(n, mean = 0, sd = sqrt(3))
  y_sample <- beta_0 + beta_1*x_sample + e_sample

  plot(x_sample, y_sample)
  return(y_sample)
}
```

```r
generate_scatter_norm()
```

```r
generate_scatter_unif <- function(n=40, beta_0 = 5, beta_1 = -2) {
  e_sample <- runif(n, min=-3, max = 3)
  y_sample <- beta_0 + beta_1*x_sample + e_sample

  plot(x_sample, y_sample)
  return(y_sample)
}
```

```r
generate_scatter_unif()
```
