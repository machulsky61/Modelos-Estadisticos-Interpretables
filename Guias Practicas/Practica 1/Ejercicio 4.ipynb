```r
aa <- "dino"

plot(datasaurus_dozen$x[datasaurus_dozen$dataset == aa],
datasaurus_dozen$y[datasaurus_dozen$dataset == aa],
main = c("datasaurus_dozen ", paste(aa)), xlab = "x", ylab = "y", pch = 20)

x_values <- datasaurus_dozen$x[datasaurus_dozen$dataset == aa]
y_values <- datasaurus_dozen$y[datasaurus_dozen$dataset == aa]

print(paste("La media de x es es", round(mean(x_values),2), " y el desvio es", round(sd(x_values),2)))
print(paste("La media de y es es", round(mean(y_values),2), " y el desvio es", round(sd(y_values),2)))
print(paste("La correlación entre ambas es", round(cor(x_values,y_values),2)))
```
